## 1. <a id="1"> Problem Statement</a>

[Back to TOC](#0)


### Business Problem
- A bike-sharing system is a service in which bikes are made available for shared use to individuals on a short term basis for a price or free. 
- Bikes are borrowed through computer-controlled docks and returned back to another dock.
- Revenue dips in US bike sharing provider due to Corona.Business plan is to accelerate the revenue as soon as ongoing lockdown come to an  end.
- Understand the factors affecting the demamd of the shared bikes in the American market
    - Identify Vartiables significant in predicting the demand of shared bikes
    - How well the variables describe the bike demands
    
### Goal 
- Linear regression model  of Demand of Shared Bikes with the available Independent Variables.
- Understand Demands variation with different featuires.
- Manipulation of business strategy to meet the demand levels and expectations of the customer.
- Understand Demand Dynamics of a new Market.

- Target Variable: cnt
- Predictors: holiday, weather, workingday,temp ...etc


### Dataset    
Based on various meteorological surveys and people's styles, the service provider firm has gathered a large dataset on daily bike demands across the American market based on some factors. 

day.csv have the following fields:
	
	- instant: record index
	- dteday : date
	- season : season (1:spring, 2:summer, 3:fall, 4:winter)
	- yr : year (0: 2018, 1:2019)
	- mnth : month ( 1 to 12)
	- holiday : weather day is a holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)
	- weekday : day of the week
	- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
	+ weathersit : 
		- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
		- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
		- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
		- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
	- temp : temperature in Celsius
	- atemp: feeling temperature in Celsius
	- hum: humidity
	- windspeed: wind speed
	- casual: count of casual users who have made a rental
	- registered: count of registered users who have made a booking on a given day
	- cnt: count of total rental bikes including both casual and registered.The model should be built taking this 'cnt' as the target variable.
    
### Additional Points
- 'weathersit' and 'season' have values as 1, 2, 3, 4 does not mean they are ordinal categorical variables. Either convert to string variables or handle it differently.(through dummy variable)
- the 'cnt' variable indicates the total number of bike rentals, including both casual and registered. The model should be built taking this 'cnt' as the target variable.
- R2_Score on testset required

## <a id="2.2"> Data Cleaning and manipulations based on EDA </a>

- The casual and registered are not available for new data hence these data are dropped.
- Drop 'atemp','instant','dteday','casual','registered'
- Add dummy variable for 'weathersit','season' ...
- One hot encoding for catefgorical variables 'weathersit','season','mnth','weekday'..etc

[Back to TOC](#0)

```python
## Temporary Call (Comment Following when running in main mode)
vehdmd = pd.read_csv('day.csv')

## Drop unnecessary Columns
vehdmd.drop(columns = [ 'atemp', 'instant', 'dteday'], inplace = True)

## Provide Season and Weathersit a Categorical Mapping as defined in dictionary
vehdmd['season_cat'] = vehdmd.season.map({1:'spring', 2: 'summer' , 3: 'fall' , 4:'winter'})

## Weathesit Mapping
##    - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
##    - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
##    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
##    - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
vehdmd['weathersit_cat'] = vehdmd.weathersit.map({1:'weatsit_1clear', 2: 'weatsit_2cloudy' , 3: 'weatsit_3lightrain' , 4:'weatsit_4heavyrain'})

## month Mapping
vehdmd['mnth_cat'] = vehdmd.mnth.map({ 1:'m1' , 2: 'm2' , 3:'m3', 4:'m4', 5:'m5' , 6:'m6' , 7:'m7'
                                            , 8:'m8' , 9:'m9' , 10:'m10', 11:'m11',12:'m12'})

## week Mapping
vehdmd['weekday_cat'] = vehdmd.weekday.map({ 1:'w1' , 2: 'w2' , 3:'w3', 4:'w4', 5:'w5' , 6:'w6' , 0:'w0'})


## Create dummy Variables for Weather and Season (One hot encoding)
status_weat = pd.get_dummies(vehdmd['weathersit_cat'],drop_first = True)
status_sson = pd.get_dummies(vehdmd['season_cat'],drop_first = True)
status_mnth = pd.get_dummies(vehdmd['mnth_cat'],drop_first = True)
status_wkdy = pd.get_dummies(vehdmd['weekday_cat'],drop_first = True)

status_weat.value_counts()
print(vehdmd['season_cat'].value_counts())
print(vehdmd['weathersit_cat'].value_counts())
#status.info()
## Concatenating the Status Data to the vehdmd
vehdmd = pd.concat([vehdmd, status_weat], axis = 1)
vehdmd = pd.concat([vehdmd, status_sson], axis = 1)
vehdmd = pd.concat([vehdmd, status_mnth], axis = 1)
vehdmd = pd.concat([vehdmd, status_wkdy], axis = 1)


## Drop the temporary columns after Dummy Variable Creation
vehdmd.drop(columns = ['season_cat', 'weathersit_cat','season','weathersit','mnth_cat','weekday_cat'], inplace = True)
vehdmd.head()

## Drop the 'Casual and Registered'
vehdmd.drop(columns = ['casual', 'registered'], inplace = True)
vehdmd.head()

```

<div class="alert alert-block alert-danger">
<b>Important Observation/Limitations :</b> The current dataset weathersituation data does not have any data for the '4:HeavyRain' data value. Hence if any new data has this attribute the Model needs to be retrained.
</div>



```python
if (EnDummyEncForYr==1):
    vehdmd['yr_cat'] = vehdmd.yr.map({0:'yr_2018', 1: 'yr_2019'})
    ## The 2018 and 2019 both are considered and the base is for other years so drop_first=True not enabled.
    status_yr = pd.get_dummies(vehdmd['yr_cat'],drop_first=True)
    vehdmd = pd.concat([vehdmd, status_yr], axis = 1)
    vehdmd.drop(columns = ['yr_cat'], inplace = True)
    vehdmd.head()
```
